## import

```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
url = 'https://www.scrapethissite.com/pages/simple/'
imdb_response = requests.get(url)
soup = BeautifulSoup(imdb_response.text, 'html.parser')
```

```python
country_info = soup.find_all('div', {'class': 'country'})
```

## Extract data




```python
country_names = []
capitals = []
populations = []
areas = []

for country in country_info:
    country_name = country.find('h3').text.strip()
    capital = country.find('span', {'class': 'country-capital'}).text
    population = country.find('span', {'class': 'country-population'}).text
    area = country.find('span', {'class': 'country-area'}).text

    country_names.append(country_name)
    capitals.append(capital)
    populations.append(population)
    areas.append(area)
```

## Create dataframe




```python
country_data = {
    'Country': country_names,
    'Capital': capitals,
    'Population': populations,
    'Area': areas
}
df_countries = pd.DataFrame(country_data)
```
